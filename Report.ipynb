## Data
For this section we will talk about for each of the two models seperately, as different data sets were used.

Our first dataset, sourced from game_statistics_export.csv, covers 453 games across 196 teams from the 2025–26 season. Data collection began with Middlebury's full season, then expanded to include every opponent Middlebury faced: Skidmore, Union College, Stevens, Plattsburgh State, Rensselaer Polytechnic Institute, University of Rochester, Carnegie Mellon, Wheaton College, Medgar Evers, Lasell University, Worcester Polytechnic Institute, Trinity College, Wesleyan University, Connecticut College, Tufts University, Keene State, Williams College, The City College of New York, Bates College, Vermont State University Johnson, Bowdoin College, Colby College, Amherst College, and Hamilton College.
Data was collected from the Middlebury Athletics page and the athletics pages of opponent programs hosted on the SIDEARM Sports platform. Using a custom JavaScript parser, we scraped play-by-play data for each game, generated lineup sequences from that data, and tracked game statistics within each lineup's context. These were then aggregated into game-level rows containing box score statistics and additional derived metrics.
For the clustering model, we selected the following eight features, each chosen for its ability to capture a distinct dimension of team identity:

**Pace** — Estimates the number of possessions per game. Higher pace indicates a preference for transition offense and a higher volume of shot attempts.
**Three-Point Attempt Rate** — The proportion of field goal attempts taken from three-point range. Higher rates suggest a perimeter-oriented offense; lower rates indicate a preference for interior play.

**Effective Field Goal Percentage** (eFG%) — Adjusts standard FG% to account for the added value of three-point makes, providing a more accurate measure of shooting efficiency.

**Free Throw Rate** — The ratio of free throw attempts to field goal attempts. Higher values suggest teams that frequently attack the basket and draw fouls.

**Turnover Percentage** (TOV%) — The share of possessions ending in a turnover. Lower values indicate better ball security.

**Rebound Rate** — The proportion of available rebounds a team secures. Strong rebounding can be a core part of a team's identity, particularly when it generates second-chance opportunities.

**Offensive Rebound Rate** — A subset of rebound rate focusing specifically on offensive boards, which directly create second-chance scoring opportunities.

**Steal Rate** — The frequency with which a team generates steals, generally reflective of active perimeter defense and defensive pressure.


Our second dataset, sourced from all_games.csv, covers 94 games that Middlebury college played across the last four seasons against the aforementioned teams and a few more. In some ways it presents more issues with independance than the other dataset-- Middlebury players are not only represented across games within the same season but additionally in games across seasons-- but it ultimately makes dependence less of an issue for away teams across players. It also contains the team totals, which the logistic regression models ended up using because that adheres more strongly to the work outlined by Teramoto & Cross (2010) and experiements to use player level stats were not successful. Data was, again, collected from the Middlebury Athletics page and the athletics pages of opponent programs hosted on the SIDEARM Sports platform. Using a custom R scraper built with rvest, we scraped boxscore data for each game from the Middlebury Athletics website, parsed raw shooting columns from made-attempt strings into percentages, split rebound totals into offensive and defensive components, and extracted game outcomes and dates from the page HTML. These were then aggregated into a flat dataframe containing team-level box score statistics for both Middlebury and their opponent in each game. After first trying the four main metrics that Teramoto & Cross (2010) point towards (field goal percentage, turnovers, rebounds, free throw percentage) we tried different different sets of features, ultimtately finding more success the more features we added, settling on:

**Rebounds** — Total rebounds secured in a game, capturing a team's overall dominance on the glass across both ends of the floor.

**Personal Fouls** — Total fouls committed. Higher values may indicate aggressive defense or poor defensive discipline, and can influence opponent free throw opportunities.

**Assists** — Total assists, reflecting the degree of ball movement and team-oriented offense. Higher assist totals generally indicate a more collaborative offensive system.

**Turnovers** — Total turnovers committed. Fewer turnovers indicate better ball security and limit opponent transition opportunities.

**Blocks** — Total blocked shots, reflecting interior defensive presence and rim protection.

**Steals** — Total steals generated, generally reflective of active perimeter defense and defensive pressure.

**Points** — Total points scored, the most direct measure of offensive output in a given game.

**Offensive Rebounds** — Offensive boards secured, which directly create second-chance scoring opportunities and extend possessions.

**Defensive Rebounds** — Defensive boards secured, limiting opponent second-chance opportunities and initiating transition offense.

**Field Goal Percentage** — The proportion of field goal attempts converted, providing a baseline measure of shooting efficiency.

**Three-Point Percentage** — Shooting efficiency from beyond the arc, reflecting perimeter scoring ability.

**Free Throw Percentage** — The proportion of free throw attempts converted, capturing a team's ability to capitalize on foul-drawing opportunities.

## Your Approach
*For the Guassian Mixture Model:* Because Gaussian Mixture Models were not covered in this course, we used Gemini to help us deconstruct the GMM implementation available in scikit-learn. This produced a CustomGaussianMixtureModel class and a CustomStandardScaler class, which we used without modification — the majority of our development time was spent on data wrangling and designing an output pipeline that could yield interpretable, analyzable results.
For visualization, we initially generated KDE covariance ellipse plots, a standard approach in clustering research. However, because these are two-dimensional projections of an eight-feature space, they were insufficient for capturing the full structure of our clusters and were ultimately not informative on their own.
To determine the optimal number of clusters, we tested a range from 2 to 9. Our starting point was guided by BIC score analysis, a common approach in the literature, but BIC consistently recommended only 1–2 clusters — a result that, upon inspection, amounted to little more than separating strong teams from weak ones. This was too shallow for our purposes.
We therefore developed a multi-criteria evaluation framework using three custom metrics:
Distribution Score — The standard deviation of the percentage of teams assigned to each cluster. A lower score indicates that teams are more evenly spread across clusters, which we prioritized to avoid one or two dominant clusters with a handful of outliers.
Log-Likelihood — A measure of how well the model fits the data. Given known limitations in our dataset, we did not expect a perfect fit, but used this metric to flag configurations that were clearly underperforming.
Average Entropy — A measure of cluster overlap. Lower entropy indicates more distinct cluster assignments, which we treated as the primary criterion since the goal of this experiment was to identify meaningfully differentiated playstyles.
Evaluating across all three metrics, we selected 7 clusters as the optimal configuration — the solution that minimized overlap while maintaining acceptable log-likelihood and distribution scores.

*For the Logistic Regression Model:* Because binary logistic (and other types of logistic)regression were covered in this course, we relied heavlily upon lectures 6-9 to build the model logistic regression model: this entails running gradient descent to optimize for binary cross entropy. After getting the nuts and bolts of the model running, we implemented a modified version of the kernel expansion function seen in class but compensated for the higher dimensionality.

Since we knew the classes were going to be imbalanced-- and that our dataset for logistic regression model was going to be quite small, evaluating was expected to pose a chalenge. Accuracy alone is insufficient when classes are imbalanced — a classifier that always predicts the majority class achieves a non-trivial baseline without learning anything. Reporting train and test accuracy together additionally allows us to diagnose overfitting. The confusion matrix, TPR, FPR, precision, and F1 decompose errors more granularly, distinguishing between the cost of false positives and false negatives. The ROC curve and AUC explore evaluation across all decision thresholds rather than fixing one at 0.5, giving a threshold-independent summary of the model's discriminative ability.

## Logistic Regression Demo

```python
import sys
import torch
import pandas as pd
import matplotlib.pyplot as plt

sys.path.append('../src')

try:
    from logistic_regression import (
        clean_df, prepare_features, BinaryLogisticRegression,
        GradientDescentOptimizer, binary_cross_entropy,
        confusion_matrix, roc_curve,
    )
    df = pd.read_csv('../data/all_games.csv')
    df = clean_df(df)

    X_train, y_train, X_test, y_test = prepare_features(df, num_kernels=3)

    model = BinaryLogisticRegression(n_features=X_train.shape[1])
    opt   = GradientDescentOptimizer(model=model, lr=0.1)

    loss_history = []
    for epoch in range(100000):
        q    = model.forward(X_train)
        loss = binary_cross_entropy(q, y_train)
        loss_history.append(loss.item())
        opt.step(X_train, y_train)
        if len(loss_history) > 1 and abs(loss_history[-1] - loss_history[-2]) < 1e-7:
            print(f"Converged at epoch {epoch}")
            break

    q_train = model.forward(X_train)
    q_test  = model.forward(X_test)

    tau = 0.5
    print(f"Best threshold (F1-optimal): {tau:.2f}")

    train_acc    = ((q_train >= tau).float() == y_train).float().mean().item()
    test_acc     = ((q_test  >= tau).float() == y_test ).float().mean().item()
    baseline_acc = max(y_test.mean().item(), 1 - y_test.mean().item())

    print(f"Train accuracy : {train_acc:.4f}")
    print(f"Test accuracy  : {test_acc:.4f}")
    print(f"Baseline acc   : {baseline_acc:.4f}")

    y_pred = (q_test >= tau).float()
    cm = confusion_matrix(y_test, y_pred)
    print("\nConfusion Matrix (rows=actual, cols=predicted):")
    print(cm)

    TN = cm[0, 0].item(); FP = cm[0, 1].item()
    FN = cm[1, 0].item(); TP = cm[1, 1].item()

    tpr       = TP / (TP + FN) if (TP + FN) > 0 else float('nan')
    fpr       = FP / (FP + TN) if (FP + TN) > 0 else float('nan')
    precision = TP / (TP + FP) if (TP + FP) > 0 else float('nan')
    recall    = tpr
    f1        = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else float('nan')

    print(f"\nTrue Positive Rate (Recall) : {tpr:.4f}")
    print(f"False Positive Rate         : {fpr:.4f}")
    print(f"Precision                   : {precision:.4f}")
    print(f"F1 Score                    : {f1:.4f}")

    fig, axes = plt.subplots(1, 2, figsize=(13, 5))

    axes[0].plot(loss_history)
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Binary Cross-Entropy Loss')
    axes[0].set_title('Training Loss')

    fpr_train, tpr_train = roc_curve(y_train, q_train)
    fpr_test,  tpr_test  = roc_curve(y_test,  q_test)
    auc_test = torch.trapz(torch.tensor(tpr_test), torch.tensor(fpr_test)).abs().item()

    axes[1].plot(fpr_train, tpr_train, label='Train')
    axes[1].plot(fpr_test,  tpr_test,  label=f'Test (AUC={auc_test:.2f})')
    axes[1].plot([0, 1], [0, 1], 'k--', label='Random')
    axes[1].set_xlabel('False Positive Rate')
    axes[1].set_ylabel('True Positive Rate')
    axes[1].set_title('ROC Curve - Win Prediction')
    axes[1].legend()

    plt.tight_layout()
    plt.show()

except FileNotFoundError as e:
    print(f"Error: Could not find a required file. Please ensure 'all_games.csv' is in the data directory and 'logistic_regression.py' is in src. Details: {e}")
except Exception as e:
    print(f"Error running logistic regression demo: {e}")
```

## Results

For the Guassian Mixure Model, to interpret our clusters, we conducted brief research into the dominant playstyles in college basketball during the 2025–26 season. A recent ESPN article notes a shift toward "big lineups" — teams averaging 6'4" or taller across their starting five — as a response to the rising offensive baseline across all levels of the sport. As shooting ability has become more universal, teams are increasingly seeking competitive advantages through size and physicality rather than perimeter creation alone. We used this context as a qualitative lens when interpreting our cluster outputs.
To visualize our results, we generated a heatmap of normalized features across all clusters, which allowed us to compare each cluster's statistical profile relative to the rest of the sample. The seven clusters and their interpretations are as follows:


*   **Cluster 0 — Physical Lineups:** Near-average performance across most categories, with shooting efficiency as a notable weakness. This cluster distinguishes itself through above-average steal rate and offensive rebounding, suggesting teams with length and physicality that generate second-chance opportunities and disrupt opposing ball-handlers.
*   **Cluster 1 — Fast but Not Rushed:** The lowest turnover percentage of any cluster by a considerable margin, indicating elite ball security. These teams also play at an above-average pace, suggesting a controlled transition style — pushing tempo without sacrificing possession.
*   **Cluster 2 — Efficient Shooting & Second-Chance Points:** Defined by above-average shooting efficiency, strong pace, high steal rate, and strong rebounding. These teams combine physicality with offensive execution, generating second-chance opportunities while scoring efficiently.
*   **Cluster 3 — Methodical Slashers:** Characterized by solid shooting efficiency and the highest free throw rate among all clusters. Their below-average pace suggests a reliance on structured half-court offense — back-cuts, post actions, and drive-and-kick sequences — that consistently generates high-quality looks and draws fouls.
*   **Cluster 4 — The Baseline:** The most statistically balanced cluster, with no pronounced strengths or weaknesses across any feature. These teams represent a league-average playstyle with no clearly defined offensive or defensive identity.
*   **Cluster 5 — Perimeter Volatility:** Attempts the highest volume of three-pointers of any cluster, but at below-average efficiency. The reliance on perimeter shot selection naturally suppresses offensive rebounding, as fewer misses land near the basket, limiting second-chance opportunities.
*   **Cluster 6 — Transition Aggressors:** The fastest-paced cluster, paired with a high rate of opponent turnovers. These teams build their offense around transition opportunities rather than three-point shooting or foul-drawing, relying on fast-break execution and disciplined decision-making in open-court situations.

For the logistic regression model, the logistic regression model converged after roughly 14,000 epochs and achieved a train accuracy of 72%, but generalized poorly to the test set at 47% — below the majority-class baseline of 61%. This is overfitting. Issues like this were expected given the small sample size of  188 (94 games * two teams a game). To really take a moment on this sentiment:

With only 188 team-game observations, finding the right model complexity is genuinely difficult. A simpler model with fewer features, no kernel expansion underfits and fails to capture the nonlinear structure in the data. A more complex model with more kernels and more features has enough capacity to memorize the training set but generalizes poorly. **The small sample size leaves almost no room between these two failure modes**, and the train/test split exacerbates this further: with only **38 test observations**, a single misclassified game swings test accuracy by over 2 percentage points, making it hard to draw reliable conclusions about which model configuration actually generalizes best. This is ultimately a data problem, and the most meaningful improvement would come from expanding the dataset to cover more teams and seasons rather than further tuning the model.

At the F1-optimal threshold of 0.5, the model correctly identified wins at a recall of 61%, but at the cost of a high false positive rate of 73%, indicating it predicted wins too liberally. The F1 score of 0.58 basically makes this model marginally better than trivial. I mean it's below the baseline.

One interesting thing to note, performance improved (albiet modestly) when RBF kernel expansion was introduced relative to the linear baseline, suggesting that the relationship between box score statistics and win outcomes is not purely linear — consistent with findings in the broader basketball analytics literature. Similarly, including the full feature set outperformed models restricted to fewer predictors. It's interesting that this was the case and that the four predictors Teramoto & Cross (2010) employed didn't do as well, though, as with most of the issues with this model, I'm sure it's tied to data limitations.
